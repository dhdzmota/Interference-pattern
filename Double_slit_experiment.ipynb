```python
#Experimento de la doble Rendija

import numpy as np
import matplotlib.pyplot as plt
from numpy import random as rd
import matplotlib.text   as tx

%matplotlib inline
```

```python
#LÃ­mites y factores del sistema:
a=10
b=-10
Num=100000
x=np.linspace(b,a,Num)
h=1.05457e-34
m=9.1093e-31
```

```python
#PartÃ­cula 1:

#PÃ¡rametros
t=0
x0=0
sigma=1
k0=1

#FunciÃ³n: 
Y1=(2*np.pi*sigma**2)**(-1/4)*sigma/(np.sqrt(sigma**2+1j*h*t/(2*m*sigma)))*np.exp(-(x-x0-2j*sigma**2*k0)**2/(4*(np.sqrt(sigma**2+1j*h*t/(2*m*sigma)))**2)-sigma**2*k0**2+1j*k0*x)

#Componentes reales e imaginarios
reY1=np.real(Y1)
imY1=np.imag(Y1)

#Graficar:
plt.title("FunciÃ³n de onda: $\Psi_1(x,t)$")
plt.plot(x,reY1,label=" Real")
plt.plot(x,imY1,label=" Imaginaria")

yy=reY1**2+imY1**2
#plt.plot(x,yy,label=" Prob")

plt.xlabel("PosiciÃ³n $x$")
plt.ylabel("$\Psi_1(x,t)$")
plt.legend()
plt.show()
```

```python
#ParÃ¡metros
t=0
x0=0
sigma=2.0
k0=8

#FunciÃ³n:
Y2=(2*np.pi*sigma**2)**(-1/4)*sigma/(np.sqrt(sigma**2+1j*h*t/(2*m*sigma)))*np.exp(-(x-x0-2j*sigma**2*k0)**2/(4*(np.sqrt(sigma**2+1j*h*t/(2*m*sigma)))**2)-sigma**2*k0**2+1j*k0*x)

#Componentes imaginarios y reales
imY2=np.imag(Y2)
reY2=np.real(Y2)

#Graficar:
plt.title("FunciÃ³n de onda: $\Psi_2(x,t)$")
plt.plot(x,reY2,label="Real")
plt.plot(x,imY2,label="Imaginaria")
plt.xlabel("PosiciÃ³n $x$")
plt.ylabel("$\Psi_2(x,t)$")
plt.legend()
plt.show()
```

```python
#Suma de ambas funciones de onda: 

#Checar si existe interferencia

#Se deben normalizar:
Y3=1/np.sqrt(2)*(Y1-Y2)

#Componentes imaginarios y reales
reY3=np.real(Y3)
imY3=np.imag(Y3)

#Graficar:
plt.title("FunciÃ³n de Onda: $\Psi_3(x,t)=\Psi_1-\Psi_2$")
plt.plot(x,reY3,label="Real")
plt.plot(x,imY3,label="Imaginaria")
plt.xlabel("PosiciÃ³n $x$")
plt.ylabel("$\Psi_3(x,t)$")
plt.legend()
plt.show()
```

```python
#Probabilidad de ambas funciones de onda:

#Generar un patrÃ³n de interferencia
YP=(np.real(Y3)**2+np.imag(Y3)**2)
plt.plot(x,YP)

#Determinar Ã¡rea debajo de la curva
A=(YP)*((a-b)/len(x))

#Graficar:
plt.title("FunciÃ³n de Probabilidad")
plt.plot(x,YP)#,label="$|\Psi_3(x,t)|^2$")
plt.xlabel("PosiciÃ³n $x$")
plt.ylabel("$|\Psi_3(x,t)|^2$")

plt.show()

#Completar la suma de 1 (en caso de ser menor a 1)
c=1-np.sum(A)
A[int(Num/2)]=A[(int(Num/2))]+c
```

```python
#GeneraciÃ³n de experimento:

#Vector entre -10 a 10
w=np.linspace(b,a,Num)
print(w)

#ObtenciÃ³n de valor aleatorio con probabilidad asignada por la funciÃ³n de onda de valor -10 a 10, 
d=rd.choice(w,Num,p=A)

y=np.linspace(0,.46,Num)

plt.title("Experimento de la partÃ­cula")
plt.xlabel("PosiciÃ³n $x$")
plt.scatter(d,y, s=.01)
plt.show()
```
